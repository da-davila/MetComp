```python
%pylab inline
import scipy as sp
from pandas import *
```

```python
rdata = loadtxt('./sparse_sample_monthrg.dat')
```

```python
rdata[rdata == -99] = nan
```

```python
data = DataFrame(rdata , columns = ['aaaa', 'mm', 'nd','nmanchas'])
```

```python
tiempo = (data.aaaa - 1610)*12 + data.mm #pasamos a meses y el mes 1 es enero de 1610 
```

```python
dataf = DataFrame([tiempo, data.nmanchas])
dataf = dataf.T
dataf = dataf.rename(columns = {dataf.columns[0] : 'tiempo'})
dataf = dataf.rename(index = dict(zip(dataf.index, dataf.tiempo)))
```

```python
t = range(1,int(max(dataf.tiempo))+1)
```

```python
dataTemp = DataFrame(t,index=t)
```

```python
dataf = dataf.join( dataTemp, how = 'outer')
```

```python
dataf.tiempo = dataf.iloc[:,2]
```

```python
dataf = dataf.drop(dataf.columns[2], 1)
```
