```python
%pylab inline
import numpy as np
import scikits.audiolab as audio
from scipy.fftpack import fft, fftfreq

```

```python
input_signal, sampling_rate, enc = audio.wavread("grabacionVad.wav")

print (input_signal[0:5]), sampling_rate, enc
```

```python
time_array = arange(0, len(input_signal)/float(sampling_rate), 1/float(sampling_rate))
```

```python
plot(time_array[0:4000], input_signal[0:4000])
xlabel("time(s)", fontsize=20)
ylabel("Amplitude", fontsize=20)
```

```python
n = len(time_array) # number of point in the whole interval
f = 200.0 #  frequency in Hz
dt = 1 / (f * n/7. ) #32 samples per frequency
```

```python
TransFou = fft(input_signal)/len(input_signal)
freq = fftfreq(n, dt)
amps = np.abs(TransFou)
```

```python
nuevDat = ifft(1.2*TransFou).real #hacemos la voz mas grave, para que suene como darth vader
```

```python
output_file = audio.Sndfile("darthVader.wav", 'w', audio.Format('wav'), 1, sampling_rate)
```

```python
output_file.write_frames(nuevDat)
output_file.close() 
```
