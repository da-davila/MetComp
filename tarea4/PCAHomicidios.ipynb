```python
data = open('homicidios.1990.a.2013.csv', 'Ur')
MatDatos = zeros((0,75))
for line in data:
    l = [i.strip() for i in line.split(',')]
    MatDatos = vstack([MatDatos,l])
```

```python
cols = (8,11,14,17,20,23,26,29,32,35,38,41,44,47,50,53,56,59,62,65,68,71,74)
for i in range(1123):
    for j in cols:
        if (MatDatos[i,j] == 'NA' or MatDatos[i,j] == '#DIV/0!' or MatDatos[i,j] == 'NaN'):
            MatDatos[i,j] = 0
```

```python
MatTasas = zeros((1122,23))
cols = (8,11,14,17,20,23,26,29,32,35,38,41,44,47,50,53,56,59,62,65,68,71,74)
for j in range (23):
    i = cols[j]
    MatTasas[:,j] = MatDatos[1:, i]
```

```python
MatCov = zeros((23,23))
for i in range (23):
    for j in range (23):
        suma = 0
        for m in range (1122):
            suma = suma + (MatTasas[m,i] - mean(MatTasas[:,i]))*(MatTasas[m,j] - mean(MatTasas[:,j]))
        MatCov[i,j] = suma/(23-1)
```
