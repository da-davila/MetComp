Vemos que la functión se repite cada unidad de phase, por lo cual debemos esperar que la transformada de Fourier tenga picos cada unidad de frecuencia.

# Aproximación de la curva con N componentes

```python
dataf = fft.fft(data.int)
freq = fftfreq(len(data), 0.002)
plot(freq, abs(dataf))
scatter(freq, abs(dataf), s = 40, alpha = 0.5)
title('Espectro de los datos')
xlabel('Frecuencia (unidades de fase ^-1)')
ylabel('Intensidad (unidades de intensidad)')
xlim([-5,5])
```

Hay picos cada unidad como fue anotado antes.

```python
def faprox(n, four): #función que toma los N primeros componentes de la transf. de Fourier y calcula la cotransformada sin los demás
    apf = copy(four) #las asignaciones en python no copian objetos :(
    if n== 0:
        apf[1:-1]=0
    else:
        apf[n+1:-n]=0 #ver fftshift
    return ifft(apf)
```

```python
chisq = zeros(11) #vector con los valores de \chi^2
for i in [4*j for j in xrange(0,11)]:
    figure(i/4)
    fap = abs(faprox(i,dataf))
    chisq[i/4] = sum((fap - data.int)**2/data.int)
    plot(fap)
    plot(data.int)
    title('Aproximacion con '+str(i/4)+' componentes')
    xlabel('fase')
    ylabel('intesidad')
```

```python
plot(chisq)
title('$\chi^{2}$ vs numero de componentes', fontsize=16)
xlabel('Numero de componentes N', fontsize=12)
ylabel('$\chi^{2}$', fontsize=12)
```
