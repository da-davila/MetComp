# Importar los datos

```python
data = read_csv('sparse_sample_monthrg.dat' , sep = ' ' ) #importar datos
data.columns = ['ano', 'mes', 'dias_obs', 'nmanchas'] #renombrar columnas
data['tiempo'] = data.mes + (data.ano-1610)*12 #pasar a tiempo en meses con el mes 1 en enero de 1610
data = data.drop(['dias_obs','ano' , 'mes'], axis = 1) #quitar las columnas que no vamos a usar
data.nmanchas[data.nmanchas == -99.0] = nan #convertir los -99.0 en nan
#Hacer que el tiempo esté uniformemente espaciado
tiempo_max = max(data.tiempo)
tiempo_min = min(data.tiempo)
t = range(tiempo_min, tiempo_max+1) #meses equiespaciados
data.index = data.tiempo #cambiar el índice
temp = DataFrame(t, columns = ['tiempo'], index = t) #dataframe temporal con los meses equiespaciados
data = merge(data, temp, left_index=True, right_index=True, how='outer') #crear el dataframe con los meses equiaespaciados
data = data.drop(['tiempo_x', 'tiempo_y'], axis = 1) #quitar las columnas sobrantes
data = data.interpolate(method = 'linear')#interpolar
#maravillarse
```

```python
where(isnan(data.nmanchas)) #veo que los primeros 27 no tienen nada
```

```python
data = data.iloc[28:,:] #los quitamos
plot(data.index, data)
title('Datos de actividad solar')
xlabel('tiempo (mes)')
ylabel('Actividad solar (unidades de actividad solar)')
```

# Análisis de Fourier

```python
fdata = fft.fft(data.nmanchas) #transformada de fourier
freq = fftfreq( len(fdata)) #frecuencias
plot(freq, abs(fdata))
title('Espectro de los datos')
xlabel('frecuencia ($mes^{-1}$)')
ylabel('Amplitud (unidades de actividad solar)')
xlim((-0.02, 0.02))
```

Vemos que predominan las frecuencias pequeñas (los periodos largos) cosa que concuerda con lo que observamos en la gráfica de los datos

```python
ind_maxamp = argsort(abs(fdata))[-8:] #índices de las primeras 8 amplitudes más grandes
abs(fdata[ind_maxamp])
```

```python
freq[ind_maxamp]
```

Vemos que la frecuencia no nula con la mayor amplitud es -0.00043554 mes^-1 que corresponde al un periodo de 2296 meses  (191.3 años) que corresponde  con lo observado en la gráfica de actividad solar.

```python
periodos = 1/freq[ind_maxamp]
```
